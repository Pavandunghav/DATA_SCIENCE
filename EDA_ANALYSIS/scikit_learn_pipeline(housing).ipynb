```python
from sklearn.pipeline import Pipeline,make_pipeline
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder

from sklearn.compose import ColumnTransformer
```

```python
tf1=ColumnTransformer(transformers=[('imputer',SimpleImputer(strategy="mean"),['Income'])],
                             
                                     remainder='passthrough')
                        
                        
```

```python
cols=data.columns
```

```python
data
```

```python
data.isna().sum()
```

```python

tf2=ColumnTransformer(transformers=[("encoder",OrdinalEncoder(),['Education'])],remainder="passthrough")
                       
```

```python
data=pd.DataFrame(tf2.fit_transform(data),columns=data.columns)
```

```python
data
```

```python
tf3=ColumnTransformer(transformers=[('encoder2'),LabelEncoder(),['Marital_Status']],remainder="passthrough")
```

```python
for i in ['Marital_Status','Dt_Customer']:
    
    data[i]=LabelEncoder().fit_transform(data[i])
    
    
    
    
```

```python
data
```

```python
pipe=make_pipeline(tf1,tf2,tf3)
```

```python
pipe.fit(data)
```
