```python
import requests


# Fetch the webpage content
webpage = requests.get(
    'https://internshala.com/internships/matching-preferences',
    
).text

# Print the webpage content
print(webpage)
```

```python
##Creattingg a object for the BeutifulSoup class and give to  it a website 

soup=BeautifulSoup(webpage,'html.parser') ##The second attribute help to parse the code of the page



```

```python
##Now going to present the the code we got 

print(soup.prettify())
```

```python
###If you want to particular tag in our soup object ##Will give all the lines with having this tag 

soup.find_all("h3")


for i in soup.find_all("h3"):
    print(i.text.strip())
```

```python
# Get all <h3> tags
h3_tags = soup.find_all("h3")
roles=[]
# Extract text from each <h3> tag
for tag in h3_tags:
    print(tag.text.strip())  # Use .strip() to remove extra whitespace
    roles.append(tag.text.strip())
    
```

```python
pd.DataFrame(roles,columns=["Roles"])
```

```python
companies=[]

for i in soup.find_all('p',class_='company-name'):
    print(i.text.strip())
    companies.append(i.text.strip())
```

```python
companies=pd.DataFrame(companies,columns=["Companies"])
companies
```
